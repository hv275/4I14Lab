```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
# Load all the clean data entries in
data_files = os.listdir("CleanData")
data = {name:np.loadtxt(f"CleanData/{name}") for name in data_files}
```

```python
# build sets of files for comparisons where such can be made for similar motions
# this code simply draws nice figures for a collection of six graphs
flex_wrist_set = ["1-WristFlex","B1-Flex","C1-WristFlex","D1-Flex","E1-Flex","F2-Flex","G1-Flex","H1-Flex"]
labels = ["Original Placement", "Reference Electrode on Back of Thumb", "Reference Electrode on Back of Wrist", "Reference Electrode on Front of Wrist Joint", "Reference Electrode on Back of Forearm", "Perpendicular Placement of Reading Electrodes", "Half electrode covered by paper", "Electrodes on back of arm"]
fig, ax = plt.subplots(2,4, figsize = (20,10))
for i, run in enumerate(flex_wrist_set):
    ax[i%2][i//2].plot(data[f"{run}.csv"])
    ax[i%2][i//2].set_title(labels[i])
    print(i%2,i//2)
plt.show()
```
